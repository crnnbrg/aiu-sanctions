# AIU global list of ineligible persons

```python
import pandas as pd
occurrence_per_nationality = pd.read_csv('data/occurrence_per_nationality.csv')
```

## Occurrences per nationality

### 1 January 2026

```python
df = occurrence_per_nationality
df[df["count"] >= 10].sort_values("count", ascending=False)[["country_name", "count"]]
```

### 1 September 2025

```python
df = pd.read_csv('data/1_sep_2025_occurrence_per_nationality.csv')
df[df["count"] >= 10].sort_values("count", ascending=False)[["country_name", "count"]]
```

## Number of countries per continent

```python
df = occurrence_per_nationality
countries_per_continent = df.groupby('continent').size().sort_values(ascending=False)
print(countries_per_continent)
```

## Number of occurences per continent

```python
df = occurrence_per_nationality
people_per_continent = df.groupby('continent')['count'].sum().sort_values(ascending=False)
print(people_per_continent)
```
